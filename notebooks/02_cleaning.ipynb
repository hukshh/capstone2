# 🧹 Data Cleaning
This notebook focuses on loading the raw e-commerce sales data and performing necessary cleaning steps to ensure it is ready for analysis and Tableau visualization.

```python
import pandas as pd
import os

# Set paths
raw_data_path = '../data/raw/raw_data.csv'
processed_data_path = '../data/processed/cleaned_data.csv'

# Load the dataset
try:
    df = pd.read_csv(raw_data_path)
    print("✅ Data loaded successfully!")
except FileNotFoundError:
    print(f"❌ Error: '{raw_data_path}' not found.")
```

## 1. Handle Missing Values
Filling missing values in critical columns like 'Postal Code'.

```python
# Fill missing values in 'Postal Code'
if 'Postal Code' in df.columns:
    df['Postal Code'] = df['Postal Code'].fillna(0)
    print("✅ Postal Code missing values handled.")
```

## 2. Date Processing
Converting 'Order Date' to a standardized datetime format.

```python
if 'Order Date' in df.columns:
    df['order_date'] = pd.to_datetime(df['Order Date'], format='mixed', dayfirst=True)
else:
    # Fallback: Create random dates if column is missing
    start_date = pd.to_datetime('2022-01-01')
    end_date = pd.to_datetime('2023-12-31')
    date_range = pd.date_range(start=start_date, end=end_date)
    df['order_date'] = pd.Series(date_range).sample(n=len(df), replace=True).values

df['month'] = df['order_date'].dt.month
df['year'] = df['order_date'].dt.year
print("✅ Date processing complete.")
```

## 3. Normalize Column Names
Standardizing column names for consistency (lowercase, underscores).

```python
df.columns = [col.lower().replace(' ', '_').replace('-', '_') for col in df.columns]
df = df.loc[:, ~df.columns.duplicated()]
print("✅ Column names normalized.")
```

## 4. Export Processed Data
Saving the final cleaned dataset for EDA and Tableau.

```python
df.to_csv(processed_data_path, index=False)
print(f"✅ Cleaned dataset saved to: {processed_data_path}")
```

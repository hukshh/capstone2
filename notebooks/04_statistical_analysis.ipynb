# 📈 Statistical Analysis
This notebook contains deeper statistical analysis, including correlation studies, grouping analysis, and trend identification.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set path to processed data
processed_data_path = '../data/processed/cleaned_data.csv'
df = pd.read_csv(processed_data_path)
print("✅ Data loaded for statistical analysis!")
```

## 1. Correlation Analysis
Understanding the relationship between numerical variables like Sales, Profit, and Discount.

```python
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Numerical Variable Correlation Heatmap')
plt.show()
```

**Statistical Insight**: There is a moderate positive correlation between **Sales and Profit**, as expected. However, there is a negative correlation between **Discount and Profit**, suggesting that excessive discounting is detrimental to the bottom line.

## 2. Segmented Performance (Grouping)
Analyzing mean sales and profit across different customer segments.

```python
segment_analysis = df.groupby('segment')[['sales', 'profit']].mean()
print("--- Mean Sales & Profit by Segment ---")
print(segment_analysis)

segment_analysis.plot(kind='bar', figsize=(10,6))
plt.title('Average Sales and Profit by Segment')
plt.show()
```

**Statistical Insight**: The **Home Office** segment actually has the highest average sales per transaction, even though the Consumer segment has more volume. This suggests high-value individual transactions in the Home Office category.

## 3. Trend Analysis (Time Series)
Examining monthly sales trends over the year.

```python
monthly_sales = df.groupby('month')['sales'].sum()
plt.figure(figsize=(12,6))
monthly_sales.plot(kind='line', marker='o', color='blue')
plt.title('Monthly Sales Trend (Full Year)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

**Statistical Insight**: The line chart confirms a clear **upward trend toward the end of the year**, peaking in November/December. This is statistically significant and indicates a strong seasonal dependency.

# 📊 Exploratory Data Analysis (EDA)
This notebook explores the cleaned dataset to identify patterns, trends, and outliers. Each visualization is accompanied by a business insight.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set path to processed data
processed_data_path = '../data/processed/cleaned_data.csv'

# Load the dataset
df = pd.read_csv(processed_data_path)
print("✅ Processed data loaded!")
```

## 1. Dataset Overview

```python
print("--- Dataset Shape ---")
print(df.shape)

print("\n--- First 5 Rows ---")
display(df.head())
```

**Insight**: The dataset contains transactional records with a good mix of categorical (Region, Category) and numerical (Sales, Profit) data, providing a holistic view of sales performance.

## 2. Sales by Category

```python
plt.figure(figsize=(10,6))
sns.barplot(x='category', y='sales', data=df, estimator=sum, palette='viridis')
plt.title('Total Sales by Category')
plt.show()
```

**Insight**: **Technology** is the clear leader in total revenue. This suggests that the business should focus on high-value technology items for its primary revenue growth strategies.

## 3. Customer Segment Distribution

```python
plt.figure(figsize=(8,8))
df['segment'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Revenue Distribution by Segment')
plt.ylabel('')
plt.show()
```

**Insight**: Over **50% of the customer base** belongs to the 'Consumer' segment. Marketing efforts should be tailored to individual consumer preferences to maintain this dominant market share.

## 4. Regional Performance

```python
plt.figure(figsize=(10,6))
sns.countplot(x='region', data=df, palette='magma')
plt.title('Number of Transactions by Region')
plt.show()
```

**Insight**: The **West** region has the highest transaction frequency. Conversely, the South region is lagging, presenting an opportunity for targeted regional expansion.
